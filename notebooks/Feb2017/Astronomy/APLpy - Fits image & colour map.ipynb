```python
from astropy.utils.data import download_file
image_file = download_file('http://data.astropy.org/tutorials/FITS-images/HorseHead.fits', cache=True)
```

```python
import aplpy
%matplotlib inline
```

```python
f = aplpy.FITSFigure(image_file,figsize=(16,9), dpi=100)
f.show_colorscale(cmap='gray')
```

```python
f = aplpy.FITSFigure(image_file,figsize=(16,9), dpi=100)
f.show_colorscale(cmap='jet')
```

```python
f = aplpy.FITSFigure(image_file,figsize=(16,9), dpi=100)
f.show_colorscale(cmap='jet')
# Contour
f.show_contour(image_file,levels=10,cmap='gray',smooth=3,linewidths=1.0)
# Title and colourbar
f.set_title('Horsehead Nebula')
f.add_colorbar()
```
