# Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-np.pi,np.pi,0.001)
plt.plot(x,np.sin(x))
```

### Add a title

```python
x = np.arange(-np.pi,np.pi,0.001)
plt.plot(x,np.sin(x))
plt.title('y = sin(x)')
```

```python
x = np.arange(-np.pi,np.pi,0.001)
plt.plot(x,np.sin(x))
plt.title(r'$y = \sin(x)$')
```

### Add x,y axis 

```python
x = np.arange(-np.pi,np.pi,0.001)
plt.plot(x,np.sin(x))
plt.title('y = sin(x)')
plt.xlabel('x (radians)')
plt.ylabel('y')
```
